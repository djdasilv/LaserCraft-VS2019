```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "output.csv"  # Change this to your actual file path
data = pd.read_csv(file_path)

# Extract columns, assuming the first row contains axis names
x = data.iloc[:, 0]  # First column for X-axis
y = data.iloc[:, 1]  # Second column for Y-axis

# Define the window size for the running average (e.g., 5)
window_size = 5
y_avg = y.rolling(window=window_size).mean()

# Plot the original data and the running average
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Original Data', marker='o', linestyle='-', alpha=0.6)
plt.plot(x, y_avg, label=f'{window_size}-point Running Average', linestyle='-', color='red', linewidth=2)

# Labels and title from CSV headers
plt.xlabel(data.columns[0])  # X-axis label from the first column header
plt.ylabel(data.columns[1])  # Y-axis label from the second column header
plt.title("CSV Data with Running Average")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from scipy.signal import find_peaks

# Load the CSV file
file_path = "data.csv"  # Change this to your actual file path
data = pd.read_csv(file_path)

# Extract columns, assuming the first row contains axis names
x = data.iloc[:, 0]  # First column for X-axis
y = data.iloc[:, 1]  # Second column for Y-axis

# Find peaks
peaks, _ = find_peaks(y)

# Get peak values and their corresponding x coordinates
peak_values = y.iloc[peaks]
peak_coordinates = x.iloc[peaks]

# Create the plot
fig = go.Figure()

# Add the original data trace
fig.add_trace(go.Scatter(x=x, y=y, mode='markers+lines', name='Original Data'))

# Add the peak points
fig.add_trace(go.Scatter(x=peak_coordinates, y=peak_values, mode='markers', 
                         marker=dict(color='red', size=10), name='Peaks'))

# Update layout with titles and axis labels
fig.update_layout(
    title="CSV Data with Peaks",
    xaxis_title=data.columns[0],  # X-axis label from the first column header
    yaxis_title=data.columns[1],  # Y-axis label from the second column header
    template="plotly_dark"  # Optional: for a dark theme, remove if not needed
)

# Show the plot
fig.show()

# Print the peak coordinates and their differences
print("Peak Coordinates:")
print(peak_coordinates)
print("\nPeak Differences (in x-coordinates):")
print(peak_coordinates.diff().dropna())
```
